```python
# scatter plot grlivarea/saleprice linear relationship
var = 'GrLivArea'
data = pd.concat([train_df['SalePrice'],train_df[var]],axis=1)
data.plot.scatter(x = var, y = 'SalePrice', ylim=(0,800000));
```

# RELATIONSHIP WITH CATEGORICAL FEATURES

```python
#box plot overallqual/saleprice  Saleprices increases as overallquality increases
var = 'OverallQual'
data = pd.concat([train_df['SalePrice'],train_df[var]],axis=1)
## plt.subplots() is a function that returns a tuple containing a figure and axes object(s). Thus when using fig, ax = plt.subplots() you unpack this tuple into the variables fig and ax. Having fig is useful if you want to change figure-level attributes or save the figure as an image file later (e.g. with fig.savefig('yourfilename.png')). You certainly don't have to use the returned figure object but many people do use it later so it's common to see. Also, all axes objects (the objects that have plotting methods), have a parent figure object anyway, thus:
## fig, ax = plt.subplots()
## is more concise than this:
##fig = plt.figure()
##ax = fig.add_subplot(111)
f, ax = plt.subplots(figsize = (8,6))
fig = sns.boxplot(x= var, y="SalePrice",data=data)
fig.axis(ymin=0,ymax=800000)
```

```python
#box plot YearBuilt/saleprice Weak tendency of prefering newer built houses
var = 'YearBuilt'
data = pd.concat([train_df['SalePrice'],train_df[var]],axis=1)
f, ax = plt.subplots(figsize = (8,6))
fig = sns.boxplot(x= var, y="SalePrice",data=data)
fig.axis(ymin=0,ymax=800000)
plt.xticks(rotation=90); ## rotate x-axis text
```

```python
## More objective approach 
## correlation matrix (heatmap style)
corrmat = train_df.corr()
f,ax = plt.subplots(figsize=(12,9))
sns.heatmap(corrmat,vmax=.8,square=True); ##squarebool, optional If True, set the Axes aspect to â€œequalâ€ so each cell will be square-shaped. 
## v max .Values to anchor the colormap, otherwise they are inferred from the data and other keyword arguments.
```

```python
#zoom in 
k = 10 # num of variables for heatmap
cols = corrmat.nlargest(k,'SalePrice')['SalePrice'].index ## nlargest(n, columns, keep='first')
cm = np.corrcoef(train_df[cols].values.T) ## .corrcoef(x, y=None, rowvar=True, bias=<no value>, ddof=<no value>)
sns.set(font_scale=1.25)## scale size of text
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()

# cbar = Whether to draw a colorbar.
# annot= If True, write the data value in each cell. If an array-like with the same shape as data, then use this to annotate the heatmap instead of the data. Note that DataFrames will match on position, not index.
# fmt = String formatting code to use when adding annotations.
# annot_kws = Keyword arguments for matplotlib.axes.Axes.text() when annot is True. aka size of text in the heatmap cell
# xticklabels, yticklabels = If True, plot the column names of the dataframe. If False, donâ€™t plot the column names. If list-like, plot these alternate labels as the xticklabels. If an integer, use the column names but plot only every n label. If â€œautoâ€, try to densely plot non-overlapping labels


## Result
# 'OverallQual', 'GrLivArea' and 'TotalBsmtSF' are strongly correlated with 'SalePrice'. 
# 'GarageCars' and 'GarageArea' are also some of the most strongly correlated variables.
# 'GarageCars' and 'GarageArea' are like twin brothers. You'll never be able to distinguish them. 
#  we just need one of these variables in our analysis (we can keep 'GarageCars' since its correlation with 'SalePrice' is higher).
# TotalBsmtSF' and '1stFloor' also seem to be twin brothers. We can keep 'TotalBsmtSF' 
```
